```python
import pandas as pd
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3],
                      'author_name': ['Тургеньев','Чехов','Островский']
                      }, columns = ['author_id','author_name'])
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                      'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 
                                     'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 
                                     'Таланты и поклонники'], 
                     'price': [450, 300, 350, 500, 450, 370, 290]
                      })
books
```

```python
author_price = pd.merge(authors, books, on='author_id', how='outer')
author_price
```

```python
top5 = author_price.sort_values(by='price').tail(5)
top5
```

```python
author_stat = author_price.groupby('author_name').agg({'price':'min'}).reset_index()
author_stat_1 = author_price.groupby('author_name').agg({'price':'max'}).reset_index()
author_stat_2 = author_price.groupby('author_name').agg({'price':'mean'}).reset_index()
author_stat = author_stat.rename(columns={'price':'min_price'})
author_stat_1 = author_stat_1.rename(columns={'price':'max_price'})
author_stat_2 = author_stat_2.rename(columns={'price':'mean_price'})
author_stat = pd.merge(author_price, author_stat, on='author_name', how='inner')
author_stat = pd.merge(author_stat, author_stat_1, on='author_name', how='inner')
author_stat = pd.merge(author_stat, author_stat_2, on='author_name', how='inner')

author_stat = author_stat.drop(['book_title', 'price', 'author_id'], axis = 1)

author_stat
```
